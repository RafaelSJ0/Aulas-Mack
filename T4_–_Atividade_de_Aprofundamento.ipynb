```python
import pandas as pd
import numpy as np
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
netflix_titles = list(uploaded.keys())[0]
df = pd.read_csv(netflix_titles)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_info = df.info()
df_descricao = df.describe()
tipo_contagem = df['type'].value_counts()
show_id_contagem = df['show_id'].value_counts()
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Contagem de Filmes e Séries')
plt.xlabel('Tipo')
plt.ylabel('Contagem')
plt.show()
df_info, df_descricao, tipo_contagem, show_id_contagem
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
monthly_added = df.resample('M', on='date_added').size()
plt.figure(figsize=(12, 6))
monthly_added.plot()
plt.title('Adição Mensal de Shows ao Netflix')
plt.xlabel('Data de Adição')
plt.ylabel('Número de Shows')
plt.show()
```
